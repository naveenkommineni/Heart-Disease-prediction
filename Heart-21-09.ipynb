```python
import pandas as pd
```

```python
hdata=pd.read_csv('heart.csv')
```

```python
hdata.head(3)
```

```python
hdata.isnull().sum()
```

```python
Thal={'fixed':3,'normal':2,'reversible':1}
```

```python
hdata['thal']=hdata['thal'].map(Thal)
```

```python
hdata.head(2)
```

```python
hdata.isnull().sum()
```

```python
hdata.thal.mean()
```

```python
hdata.thal.fillna(hdata.thal.mean(),inplace=True)
```

```python
hdata.isnull().sum()
```

```python
x=hdata.drop('target',axis=1)
```

```python
y=hdata['target']
```

```python
from sklearn import tree
```

```python
ctree=tree.DecisionTreeClassifier()
```

```python
ctree
```

```python
ctree.fit(x,y)
```

```python
from sklearn import metrics
```

```python
y_pred=ctree.predict(x)
```

```python
metrics.accuracy_score(y,y_pred)
```

```python
ctree.feature_importances_
```

```python
comp_df = pd.DataFrame()
comp_df['cols'] = x.columns
comp_df['imp'] = ctree.feature_importances_*100

comp_df
```

```python
X=hdata[['age','cp','chol','oldpeak','ca','thalach','trestbps']]
```

```python
Y=hdata['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.4,random_state=1)
```

```python
ctree.fit(X_train,Y_train)
```

```python
Y_pred=ctree.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,Y_pred)
```

```python
#Tuning
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= { "criterion"      : ['gini', 'entropy'],    
               "max_features"   : [1,2 , 4, 5 ],             
               "splitter"       : ['best', 'random'],  
               "min_samples_leaf" : [2, 11,16,  26],
               "max_depth"        : [1,26,31,46,52]
            }
```

```python
optree = GridSearchCV(ctree, param_grid, scoring = 'accuracy', verbose=2, cv=5) 
```

```python
optree.fit(X_train,Y_train)
```

```python
optree.best_params_
```

```python
optree.best_score_
```

```python
optree.best_estimator_
```

```python
Y_test_pred=optree.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,Y_test_pred)
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
```

```python
rftree=RandomForestClassifier(n_estimators=100)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.4,random_state=1)
```

```python
rftree.fit(X_train,Y_train)
```

```python
Y_rpred=rftree.predict(X_test)
```

```python
metrics.accuracy_score(Y_test,Y_rpred)
```
